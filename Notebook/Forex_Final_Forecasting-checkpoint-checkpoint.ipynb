```python
# Download daily USD/INR
df = yf.download('USDINR=X', start='2015-01-01', end='2023-12-31')

# Keep Close only; clean
df = df[['Close']].rename(columns={'Close':'USDINR_Close'}).copy()
df.index = pd.to_datetime(df.index, errors='coerce')
df = df.sort_index().dropna()
df = df.asfreq('B')                      # business days
df['USDINR_Close'] = df['USDINR_Close'].ffill()

print(df.head(), "\n\nRows:", len(df))
```

```python

split_idx = int(len(df)*0.8)
train, test = df.iloc[:split_idx].copy(), df.iloc[split_idx:].copy()
print("Train:", train.shape, "| Test:", test.shape)
```

```python
from statsmodels.tsa.arima.model import ARIMA

# Use a safe order first; you can tune later if you like
arima_order = (1,1,1)

# Fit on train
arima_model = ARIMA(train['USDINR_Close'], order=arima_order).fit()

# Predict exactly the test length
arima_pred = arima_model.forecast(steps=len(test))
# Align index to test dates
arima_pred = pd.Series(arima_pred, index=test.index).astype(float).ffill().bfill()

# Metrics
rmse_arima = sqrt(mean_squared_error(test['USDINR_Close'], arima_pred))
mae_arima  = mean_absolute_error(test['USDINR_Close'], arima_pred)
print(f"ARIMA{arima_order} ‚Üí RMSE: {rmse_arima:.4f}, MAE: {mae_arima:.4f}")

# Quick sanity check
print("NaN in ARIMA predictions:", arima_pred.isna().sum())
```

```python
from prophet import Prophet
from sklearn.metrics import mean_squared_error, mean_absolute_error
from math import sqrt

# 1. Flatten multi-index (if not already done)
train_flat = train.copy()
train_flat.columns = ['_'.join(col).strip() if isinstance(col, tuple) else col for col in train.columns]

test_flat = test.copy()
test_flat.columns = ['_'.join(col).strip() if isinstance(col, tuple) else col for col in test.columns]

# 2. Detect the target column for USDINR
target_col = [c for c in train_flat.columns if 'USDINR' in c][0]
print("‚úÖ Target column for Prophet:", target_col)

# 3. Prepare Prophet train data
train_prophet = train_flat.reset_index().rename(columns={'Date': 'ds', target_col: 'y'})
train_prophet['ds'] = pd.to_datetime(train_prophet['ds'], errors='coerce')
train_prophet = train_prophet.dropna(subset=['ds', 'y'])

# 4. Fit Prophet
prophet_model = Prophet(daily_seasonality=False, yearly_seasonality=True, weekly_seasonality=False)
prophet_model.fit(train_prophet)

# 5. Prepare Prophet test data
test_prophet = test_flat.reset_index().rename(columns={'Date': 'ds'})
test_prophet['ds'] = pd.to_datetime(test_prophet['ds'], errors='coerce')

# 6. Predict test period
forecast = prophet_model.predict(test_prophet[['ds']])
prophet_pred = pd.Series(forecast['yhat'].values, index=test_flat.index)

# 7. Evaluate
rmse_prophet = sqrt(mean_squared_error(test_flat[target_col], prophet_pred))
mae_prophet = mean_absolute_error(test_flat[target_col], prophet_pred)

print(f'‚úÖ Prophet ‚Üí RMSE: {rmse_prophet:.4f}, MAE: {mae_prophet:.4f}')
print(prophet_pred.head())

```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from sklearn.preprocessing import MinMaxScaler

# Scale full series 0-1
scaler = MinMaxScaler((0,1))
scaled_all = scaler.fit_transform(df[['USDINR_Close']])

# Split scaled by same index boundary
scaled_train = scaled_all[:split_idx]
scaled_test  = scaled_all[split_idx:]

lookback = 60  # past days used to predict next day

# Build sequences for train
X_train, y_train = [], []
for i in range(lookback, len(scaled_train)):
    X_train.append(scaled_train[i-lookback:i, 0])
    y_train.append(scaled_train[i, 0])
X_train, y_train = np.array(X_train), np.array(y_train)
X_train = X_train.reshape((X_train.shape[0], X_train.shape[1], 1))

# Model
lstm_model = Sequential([
    LSTM(50, return_sequences=True, input_shape=(lookback,1)),
    Dropout(0.2),
    LSTM(50),
    Dropout(0.2),
    Dense(1)
])
lstm_model.compile(optimizer='adam', loss='mean_squared_error')

# Train
history = lstm_model.fit(X_train, y_train, epochs=30, batch_size=32, validation_split=0.1, verbose=1)

# Build inputs for test predictions (use tail of train + test)
inputs = scaled_all[len(scaled_all) - len(scaled_test) - lookback:]
X_test = []
for i in range(lookback, len(inputs)):
    X_test.append(inputs[i-lookback:i, 0])
X_test = np.array(X_test).reshape((-1, lookback, 1))

pred_scaled = lstm_model.predict(X_test, verbose=0)
pred_lstm = scaler.inverse_transform(pred_scaled).ravel()

# Align length to test set exactly
predicted_lstm = pd.Series(pred_lstm[-len(test):], index=test.index)

# Metrics
rmse_lstm = sqrt(mean_squared_error(test['USDINR_Close'], predicted_lstm))
mae_lstm  = mean_absolute_error(test['USDINR_Close'], predicted_lstm)
print(f"LSTM     ‚Üí RMSE: {rmse_lstm:.4f}, MAE: {mae_lstm:.4f}")
```
